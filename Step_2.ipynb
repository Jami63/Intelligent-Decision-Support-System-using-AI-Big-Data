```python
import pandas as P
```

```python
D = P.read_csv("Super Market Data Clean.csv")
```

```python
print(len(D))
```

```python
print(D.columns)
```

```python
Quantity = D["Quantity"].sum ()
print(Quantity)
```

```python
Product = D["Product line"].value_counts
print(Product)
```

```python
PGQ = D.groupby ("Product line")["Quantity"].sum ()
print(PGQ)
```

```python
Date = D.groupby ("Date")["Quantity"].sum ()
print(Date)
```

```python
SV = D.groupby("Product line")["Quantity"].sum ().sort_values(ascending = False).head(3)
print(SV)
```
