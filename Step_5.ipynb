```python
import pandas as P
```

```python
D = P.read_csv ("Super Market Data.csv")
```

```python
print(D.columns)
```

```python
X = D.drop(["Quantity"] , axis = 1)
Y = D["Quantity"]
```

```python
X = P.get_dummies (X , drop_first = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_Train , X_Test , Y_Train , Y_Test = train_test_split (X , Y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
L = LinearRegression ()
D = DecisionTreeClassifier ()
R = RandomForestClassifier ()
```

```python
L.fit (X_Train , Y_Train)
```

```python
D.fit (X_Train , Y_Train)
```

```python
R.fit (X_Train , Y_Train)
```

```python
YPL = L.predict (X_Test)
YPD = D.predict (X_Test)
YPR = R.predict (X_Test)
```

```python
import numpy as N
from sklearn.metrics import mean_squared_error , r2_score , mean_absolute_error
```

```python
R2_L = r2_score (Y_Test , YPL)
R2_D = r2_score (Y_Test , YPD)
R2_R = r2_score (Y_Test , YPR)
```

```python
MSE_L = mean_squared_error (Y_Test , YPL)
MSE_D = mean_squared_error (Y_Test , YPD)
MSE_R = mean_squared_error (Y_Test , YPR)
```

```python
RMSE_L = N.sqrt(MSE_L)
RMSE_D = N.sqrt(MSE_D)
RMSE_R = N.sqrt(MSE_R)
```

```python
MAE_L = mean_absolute_error (Y_Test , YPL)
MAE_D = mean_absolute_error (Y_Test , YPD)
MAE_R = mean_absolute_error (Y_Test , YPR)
```

```python
print("Linear Regression R2 Score : " , R2_L * 100 , "%")
print("Decision Tree Classifier R2 Score : " , R2_D * 100 , "%")
print("Random Forest Classifier R2 Score : " , R2_R * 100 , "%")
```

```python
print("Linear Regression RMSE : " , RMSE_L)
print("Decision Tree Classifier RMSE : " , RMSE_D)
print("Random Forest Classifier RMSE : " , RMSE_R)
```

```python
print("Linear Regression MAE : " , MAE_L)
print("Decision Tree Classifier MAE : " , MAE_D)
print("Random Forest Classifier MAE : " , MAE_R)
```

```python
print("Linear Regression MSE : " , MSE_L)
print("Decision Tree Classifier MSE : " , MSE_D)
print("Random Forest Classifier MSE : " , MSE_R)
```

```python
Linear_Regression_Prediction = YPL[:5]
Decision_Tree_Classifier_Prediction = YPD[:5]
Random_Forest_Classifier_Prediction = YPR[:5]
```

```python
print("Linear Regression Prediction : " , Linear_Regression_Prediction)
print("Decision Tree Classifier Prediction : " , Decision_Tree_Classifier_Prediction)
print("Random Forest Classifier Prediction : " , Random_Forest_Classifier_Prediction)
```
