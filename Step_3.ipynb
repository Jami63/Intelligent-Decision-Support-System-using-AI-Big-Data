```python
import pandas as P
```

```python
D = P.read_csv("Super Market Data Clean.csv")
```

```python
import matplotlib.pyplot as PLOT
```

```python
print(D.columns)
```

```python
SUM = D.groupby("Product line")["Quantity"].sum ()
```

```python
SUM.plot(kind = "bar")
PLOT.show()
```

```python
SUM_BY_DATE = D.groupby("Date")["Quantity"].sum()
```

```python
SUM_BY_DATE.plot ()
PLOT.show ()
```

```python
SV_PRODUCT_QUANTITY = D.groupby("Product line")["Quantity"].sum().sort_values(ascending = False).head(3)
```

```python
SV_PRODUCT_QUANTITY.plot(kind = "bar")
PLOT.show()
```

```python
QUANTITY_HISTOGRAM = D["Quantity"].hist()
PLOT.show()
```
